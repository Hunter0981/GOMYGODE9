```python
1#Write a Python program for a binary search.
def binary_search(arr, target):
    left = 0
    right = len(arr) - 1

    while left <= right:
        mid = (left + right) // 2

        if arr[mid] == target:
            return True
        elif arr[mid] < target:
            left = mid + 1
        else:
            right = mid - 1

    return False

# Test cases
arr1 = [1, 2, 3, 5, 8]
target1 = 6
print(binary_search(arr1, target1))  # Output: False

arr2 = [1, 2, 3, 5, 8]
target2 = 5
print(binary_search(arr2, target2))  # Output: True
2#
def power(a, b):
    return a ** b

# Test case
result = power(3, 4)
print(result)  # Output: 81

3#
def bubble_sort(arr):
    n = len(arr)

    # Traverse through all array elements
    for i in range(n - 1):
        # Last i elements are already in place
        for j in range(0, n - i - 1):
            # Swap if the element found is greater than the next element
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]

# Test case
data = [29, 13, 22, 37, 52, 49, 46, 71, 56]
bubble_sort(data)
print(data)  # Output: [13, 22, 29, 37, 46, 49, 52, 56, 71]

4#

def merge_sort(arr):
    if len(arr) <= 1:
        return arr

    mid = len(arr) // 2
    left = arr[:mid]
    right = arr[mid:]

    left = merge_sort(left)
    right = merge_sort(right)

    return merge(left, right)


def merge(left, right):
    merged = []
    left_index = 0
    right_index = 0

    while left_index < len(left) and right_index < len(right):
        if left[left_index] < right[right_index]:
            merged.append(left[left_index])
            left_index += 1
        else:
            merged.append(right[right_index])
            right_index += 1

    # Append any remaining elements
    while left_index < len(left):
        merged.append(left[left_index])
        left_index += 1

    while right_index < len(right):
        merged.append(right[right_index])
        right_index += 1

    return merged

# Test case
data = [29, 13, 22, 37, 52, 49, 46, 71, 56]
sorted_data = merge_sort(data)
print(sorted_data)  # Output: [13, 22, 29, 37, 46, 49, 52, 56, 71]

5#
def quick_sort(arr):
    if len(arr) <= 1:
        return arr

    pivot = arr[len(arr) // 2]
    left = [x for x in arr if x < pivot]
    middle = [x for x in arr if x == pivot]
    right = [x for x in arr if x > pivot]

    return quick_sort(left) + middle + quick_sort(right)

# Test case
data = [29, 13, 22, 37, 52, 49, 46, 71, 56]
sorted_data = quick_sort(data)
print(sorted_data)  # Output: [13, 22, 29, 37, 46, 49, 52, 56, 71]



```
